4. Now write a function to group movies released after 2021 based on genres and find the average rating of each genre. 

```python
# Insert your code here.
def genres_and_average_rating(df, year = 2021):
    # Filter movies released after the specified year
    df_filtered = df[df['startYear'] > year]

    # Split genres into separate rows
    df_exploded = df_filtered.assign(genres = df_filtered['genres'].str.split('|')).explode('genres')

    # Group by genres and calculate the average rating
    genre_rating = df_exploded.groupby('genres')['averageRating'].mean().reset_index()

    # Rename columns for clarity
    genre_rating.columns = ['Genre', 'AverageRating']

    return genre_rating

genres_and_average_rating(df)
```

8. Use appropriate plots to visualize the data in the tasks below.
- Find the average ratings of movies categorized into different genres for the period from 2010 to 2015.
- Find the top 100 movies based on the average ratings and plot their ratings versus the title.
- Find the top 100 movies based on the number of voters and plot the number of voters versus the title.
- Find the numbers of adult film and non-adult films and plot them using bar charts.
- Find the top 100 movies based on the longest run time and plot the run time versus the title.

```python
# Ensure 'startYear' is numeric
df['startYear'] = pd.to_numeric(df['startYear'], errors='coerce')

# Filter movies released between 2010 and 2015
filtered_df = df[(df['startYear'] >= 2010) & (df['startYear'] <= 2015)]

# Explode genres to handle movies with multiple genres
filtered_df = filtered_df.assign(genres=filtered_df['genres'].str.split('|')).explode('genres')

# Calculate average ratings for each genre
average_ratings = filtered_df.groupby('genres')['averageRating'].mean().sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(50, 10)) 
average_ratings.plot(kind='bar', color='skyblue') 
plt.title('Average Ratings of Movies Categorized by Genres (2010-2015)') 
plt.ylabel('Genre',fontsize=12) 
plt.xlabel('Average Rating', fontsize=12) 
plt.xticks(rotation=45, ha='right', fontsize=5) # Rotate labels and align to the right to prevent overlap 
#plt.tight_layout() # Adjust layout to fit everything nicely 
plt.show()
```

```python
# Ensure 'averageRating' is numeric
df['averageRating'] = pd.to_numeric(df['averageRating'], errors='coerce')

# Sort the DataFrame by average ratings in descending order
sorted_df = df.sort_values(by='averageRating', ascending=False)

# Select the top 100 movies
top_100_df = sorted_df.head(100)

# Plot the ratings vs. titles
plt.figure(figsize=(10, 20))
plt.barh(top_100_df['primaryTitle'], top_100_df['averageRating'], color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Movie Title',fontsize=5)
plt.title('Top 100 Movies Based on Average Ratings')
plt.gca().invert_yaxis()  # To have the highest rated movie at the top
plt.show()
```

```python
# Ensure 'numVotes' is numeric
df['numVotes'] = pd.to_numeric(df['numVotes'], errors='coerce')

# Sort the DataFrame by number of voters in descending order
sorted_df = df.sort_values(by='numVotes', ascending=False)

# Select the top 100 movies
top_100_voters_df = sorted_df.head(100)

# Plot the number of voters vs. titles
plt.figure(figsize=(10, 20))
plt.barh(top_100_voters_df['primaryTitle'], top_100_voters_df['numVotes'], color='skyblue')
plt.xlabel('Number of Voters')
plt.ylabel('Movie Title')
plt.title('Top 100 Movies Based on Number of Voters')
plt.gca().invert_yaxis()  # To have the highest voted movie at the top
plt.show()
```

```python
# Count the number of adult and non-adult films
adult_counts = df['isAdult'].value_counts()

# Plot the counts using a bar chart
plt.figure(figsize=(8, 6)) 
bars = plt.bar(adult_counts.index, adult_counts.values, color=['skyblue', 'orange'], tick_label=['Non-Adult', 'Adult']) 
# Add count labels on top of the bars 
for bar in bars: 
    yval = bar.get_height() 
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom') # Adjust the position of the labels 
plt.title('Number of Adult and Non-Adult Films') 
plt.xlabel('Film Type') 
plt.ylabel('Count')
plt.show()
```

```python
# Ensure 'runtimeMinutes' is numeric
df['runtimeMinutes'] = pd.to_numeric(df['runtimeMinutes'], errors='coerce')

# Sort the DataFrame by run time in descending order
sorted_df = df.sort_values(by='runtimeMinutes', ascending=False)

# Select the top 100 movies
top_100_runtime_df = sorted_df.head(100)

# Plot the run time vs. titles
plt.figure(figsize=(10, 20))
plt.barh(top_100_runtime_df['primaryTitle'], top_100_runtime_df['runtimeMinutes'], color='skyblue')
plt.xlabel('Run Time (Minutes)')
plt.ylabel('Movie Title')
plt.title('Top 100 Movies Based on Longest Run Time')
plt.gca().invert_yaxis()  # To have the longest run time movie at the top
plt.show()
```
